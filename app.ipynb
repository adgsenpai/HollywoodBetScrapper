```python
# open hollywood.xlsx
import pandas as pd
df = pd.read_excel('betway.xlsx')
df

df2 = pd.read_excel('hollywood.xlsx')
df2

# iterate row by row make t1 and t2 uppercase
df2['t1'] = df2['t1'].str.upper()
df2['t2'] = df2['t2'].str.upper()
df2
#drop unamed 0 column
df2.drop(columns=['Unnamed: 0'], inplace=True)
df2

# iterate df and df2 row by row find where t1 and t2 are equal

bestDF = pd.DataFrame(columns=['t1','t2','o1','o2','o3','arb'])

for index, row in df.iterrows():
    for index2, row2 in df2.iterrows():
        # check if row['t1'] and row2['t1'] are similar and row['t2'] and row2['t2'] are similar
        if (row['t1'] in row2['t1']) and (row['t2'] in row2['t2']) or (row2['t1'] in row['t1']) and (row2['t2'] in row['t2']):
            #find the biggest o1,o2,o3
            o1 = max(row['o1'], row2['o1'])
            o2 = max(row['o2'], row2['o2'])
            o3 = max(row['o3'], row2['o3'])
            # tell me which bookmaker has the best odds and add to new column as for example o1 betway
            if o1 == row['o1']:
                o1b = 'betway'
            else:
                o1b = 'hollywood'
            if o2 == row['o2']:
                o2b = 'betway'
            else:
                o2b = 'hollywood'
            if o3 == row['o3']:
                o3b = 'betway'
            else:
                o3b = 'hollywood'
            # add to new dataframe

            # compute new arb
            arb = 1/o1 + 1/o2 + 1/o3
            bestDF = bestDF.append({'t1':row['t1'], 't2':row['t2'], 'o1':o1, 'o2':o2, 'o3':o3, 'arb':arb, 'o1b':o1b, 'o2b':o2b, 'o3b':o3b}, ignore_index=True)            

# sort by arb
bestDF.sort_values(by=['arb'], ascending=True)

            
            
```
